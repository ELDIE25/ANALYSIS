```python
number = 120

# For loop to find factors of 120
print("Factors of 120 using a for loop are:")
for i in range(1, number + 1):
    if number % i == 0: #mod qui represent le remainder
        print(i)

# While loop to find factors of 120
print("Factors of 120 using a while loop are:")
i = 1
while i <= number:
    if number % i == 0:
        print(i)
    i= i + 1
```

Name , Current Amount, Withdrawal Amount and Age Of Customer ..
Now Create a method,way,function to do a subtraction to get the balance of each user in the dataframe and create a column that
all the users balance after withdrawal can be found.also give the descriptive summary of the dataframe , give the total number of rows and columns in a printed manner,
return the new dataframe holding the new column which is the balance column and also perform descriptive summary on it

```python
import pandas as pd
import numpy as np

dataFrame = {
    'Name': ['Abigail', 'Peter', 'Paul', 'Mary', 'Frida','Leona','Ade','Adeline','John','Philip'],
    'Current Amount': [18470, 203980, 3034780, 4020, 2334, 872642,12323,1399843,343413,2948],
    'Withdrawal Amount':[5000, 20000, 1000000, 1000, 1500, 12000, 5000, 300000, 45000, 500],
    'Age of Customer':[23, 34, 20, 22, 36, 20, 33, 25, 31, 28]
}

df = pd.DataFrame(dataFrame)
df
```

```python
def substract (df,column1,column2):
    df['balance'] = df[column1] - df[column2]
    return df
```

```python
substract(df, 'Current Amount' ,'Withdrawal Amount')
```
